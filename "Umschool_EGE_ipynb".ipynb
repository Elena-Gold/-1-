# **СПИСКОВЫЕ ВКЛЮЧЕНИЯ**

```
a = [i**2 for i in range(1,10)]
print(a)
```

```
b = [1,2,3,4,5]
a = [i*3 for i in b]
print(a)
```

```
b = [1,2,3,4,5]
a = [i for i in b if i % 2 == 0]
print(a)
```

```
s = '1 4'
a,b = map(int, s.split())
print(a,b)
```

```
s = '1 4'
print(s.split())
a, b = s.split()
print(a+b)
```
